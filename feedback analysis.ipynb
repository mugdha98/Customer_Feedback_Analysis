```python
list_feedback=[]
list_target=[]
with open('data/en-training.txt') as develop:
    for i in develop:
        feedback = i.strip().lower().split()
        if len(feedback) >= 1:
            target = feedback[-1]
            del(feedback[0])
            del(feedback[-1])
            list_feedback.append(feedback)
            list_target.append(target)
```

```python
len(list_target),len(list_feedback)
```

```python
list_feedback
```

```python
stopwords=[]
with open('stopwords.txt')as stops:
    for i in stops:
        stopwords.append(i.strip())
stopwords
```

```python
temp_feedback=[]
for i in list_feedback:
    temp1_feedback=[]
    for j in i:
        if j not in stopwords:
            temp1_feedback.append(j)
    temp_feedback.append(temp1_feedback)
len(temp_feedback)
```

```python
len(list_target),len(temp_feedback)
```

```python
from collections import defaultdict
freq=defaultdict(int)
refined_feedback=[]
for i in temp_feedback:
    for j in i:
        freq[j]+=1
for i in temp_feedback:
    temp2_feedback=[]
    for j in i:
        if freq[j]>1:
            temp2_feedback.append(j)
    refined_feedback.append(temp2_feedback)
refined_feedback
```

```python
count=0
for i in refined_feedback:
    if len(i)==0:
        refined_feedback.remove(i)
        list_target.pop(count)
    count+=1
refined_feedback
```

```python
len(list_target),len(refined_feedback)
```

```python
import gensim 
dic=gensim.corpora.Dictionary(refined_feedback)
bow_list=[]
for i in refined_feedback:
    bow_list.append(dic.doc2bow(i))
bow_list
count=len(dic)
dense_bow=gensim.matutils.corpus2dense(bow_list,num_terms=count).transpose()
dense_bow
```

```python
target=[]
for i in list_target:
    if i=='comment':
        target.append(0)
    elif i =='complaint':
        target.append(1)
    elif i =='meaningless':
        target.append(2)
    elif i =='request':
        target.append(3)
    elif i =='bug':
        target.append(4)
    else:
        target.append(5)
# tempdic=defaultdict(int)
# for i in list_target:
#     tempdic[i]+=1
target
# tempdic
```

```python
bow_list
tfidf=gensim.models.TfidfModel(bow_list)
tfidf
```

```python
tfidf_list=tfidf[bow_list]
tfidf_list
dense_tfidf=gensim.matutils.corpus2dense(tfidf_list,num_terms=count).transpose() 
len(dense_tfidf)
dense_tfidf
```
